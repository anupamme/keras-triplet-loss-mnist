````python
#### BEGIN CODE
input_size, instances = df.shape
epochs = 200
batch_size = 4
batches_per_epoch = instances // batch_size
learning_rate = 0.0005
log_error_frequency = 20

layers = [input_size, 128, 64, 32]
activ_func = tf.nn.relu

initializer = tf.variance_scaling_initializer()

class Autoencoder:
    """Class representing a stacked autoencoder.
    """

    def __init__(self, layers, activ_func, learning_rate):
        """Creates a new autonecoder object given the specified arguments.
        
        Arguments:
            layers {A list of integer} -- Contains the input layer size and the encoder layer sizes
            activ_func {tf activation function} -- The tensorflow activation function of choice
            learning_rate {float} -- The learning rate to be used during training.
        """
        self._activ_func = activ_func
        self._encoder = self._create_encoder_net(layers)
        decode_layers = layers[::-1]
        self._decoded_output = self._create_decoder_net(decode_layers, self._encoder)
        self._create_loss()


    def _create_encoder_net(self, layers):
        """Creates the encoder graph of the network given the specified layer sizes.
        
        Arguments:
            layers {List of integers} -- Contains the input layer size and the encoder layer sizes.
        
        Returns:
            tf tensor -- Returns a tensor representing the encoders final condensed representation.
        """
        assert(isinstance(layers, list))

        self._X = tf.placeholder(tf.float32, shape=[None, layers[0]])
        encoding = self._X
        for i, layer in enumerate(layers[1:]):
            assert(isinstance(layer, int))
            b = tf.Variable(tf.zeros(layer))
            w = tf.Variable(initializer([layers[i], layer]), dtype=tf.float32)
            encoding = activ_func(tf.matmul(encoding, w) + b)
        return encoding
    

    def _create_decoder_net(self, layers, encoding):
        """Creates the decoder graph of the network given the specified layer sizes, and the encoder tensor.
        
        Arguments:
            layers {A list of integer} -- Contains the decoder layer sizes.
            encoding {tf tensor} -- A tensor representing the encoders final condensed representation.
        
        Returns:
            tf tensor -- Returns a tensor representing the decoders final final output (reconstructed input).
        """
        decoding = encoding
        for i, layer in enumerate(layers[1:-1]):
            b = tf.Variable(tf.zeros(layer))
            w = tf.Variable(initializer([layers[i], layer]), dtype=tf.float32)
            decoding = activ_func(tf.matmul(decoding, w) + b)
        # Sigmoid output
        b = tf.Variable(tf.zeros(layers[-1]))
        w = tf.Variable(initializer([layer, layers[-1]]), dtype=tf.float32)
        return tf.nn.sigmoid(tf.matmul(decoding, w) + b)
    
    
    def _create_loss(self):
        """Creates the loss tensor, the optimizer, a training tensor that attempts to minimize the loss.
        """
        self._loss = tf.reduce_mean(tf.square(self._decoded_output - self._X))
        self._optimizer = tf.train.AdamOptimizer(learning_rate)
        self._train = self._optimizer.minimize(self._loss)
    

    def get_input(self):
        """Returns the input tensor of the network. Non-fixed size.
        Returns:
            tf tensor -- Tensor for feeding input to the network. Can feed multiple instances at once.
        """
        return self._X
    
    def get_encoding(self):
        """Returns the encoders representation tensor. Depends on input size. 
        Returns:
            tf tensor -- Tensor for getting the encoders representation or feeding a representation to the decoder.
        """
        return self._encoder
    
    def get_output(self):
        """Returns the decoders representation tensor. Depends on input size.
        This is the networks final output (reconstructed input).

        Returns:
            tf tensor -- Tensor for getting the decoders output.
        """
        return self._decoded_output
    
    def get_loss(self):
        """Loss function tensor. 
        
        Returns:
            tf tensor -- Tensor for loss evaluation.
        """
        return self._loss
    
    def get_train_op(self):
        """Tensor for perfoming train updates.
        
        Returns:
            tf tensor -- Tensor for training.
        """
        return self._train


print('Training for %d epochs and %d batches per epoch' % (epochs, batches_per_epoch))
autoencoder_nnet = Autoencoder(layers, activ_func, learning_rate)
X = autoencoder_nnet.get_input()
encoder = autoencoder_nnet.get_encoding()
output_layer = autoencoder_nnet.get_output()
loss = autoencoder_nnet.get_loss()
train = autoencoder_nnet.get_train_op()

init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)    

    # Calculate initial loss
    losses = []
    for name, timeseries in df.iteritems():
        batch = np.expand_dims(timeseries, axis=0)
        batch_loss, output = sess.run([loss, output_layer], feed_dict={X: batch})
        losses.append(batch_loss)        
    print('Epoch 0 Average Initial loss = %f' % np.mean(losses))
    
    # Run training loop
    for epoch in range(1, epochs + 1, 1):
        losses = []
        for b in range(batches_per_epoch):
            batch = np.transpose(df.sample(n=batch_size, replace=False, axis=1))
            batch_loss, _, output = sess.run([loss, train, output_layer], feed_dict={X: batch})
            losses.append(batch_loss)
        if epoch % log_error_frequency == 0 or epoch == epochs:
            print('Epoch %d Average loss = %f' % (epoch, np.mean(losses)))
    
    # Create reconstructed output and a reduced rrepresentation
    outputs_val = []
    reduced_val = []
    for name, timeseries in df.iteritems():
        batch = np.expand_dims(timeseries, axis=0)
        batch_loss, output, reduced = sess.run([loss, output_layer, encoder], feed_dict={X: batch})
        outputs_val.append(output[0])
        reduced_val.append(reduced[0])
    outputs_val = np.transpose(np.array(outputs_val))
    print('\nFinished calculating reconstructed outputs.')
#### END CODE
````

````python
time_serie_col = 26
actuals = df.iloc[:, time_serie_col].values
reconstructed = outputs_val[:, time_serie_col]
##### BEGIN CODE
#calculate error metric here
error_metric = (np.square(actuals - reconstructed)).mean(axis=0)
##### END CODE
fontsize=16
print('V1: The reconstruction of time serie %d' % time_serie_col)
plt.figure(figsize=(20,10))
plt.plot(actuals, color="blue", label="ACTUAL TIME SERIE")
plt.plot(reconstructed, color="red", label="RECONSTRUCTED TIME SERIE - METRIC=%.6f" %(error_metric))
plt.title('STACKED AUTO-ENCODERS', fontsize=fontsize)
plt.xlabel('time')
plt.ylabel('levels')
plt.legend(loc='best');
plt.show()


print('V2: What is the average mean square error of the reconstructions over time.')
square_error_evolution = (np.square(df.values - outputs_val)).mean(axis=1)
plt.figure(figsize=(20, 10))
plt.plot(square_error_evolution, color="red", label="RECONSTRUCTION MEAN SQUARE ERROR")
plt.title('STACKED AUTO-ENCODERS RECONSTRUCTION MEAN SQUARE ERROR', fontsize=fontsize)
plt.xlabel('time')
plt.ylabel('mean square error')
plt.legend(loc='best');
plt.show()


np.random.seed(1)
print('V3: A 2D UMAP reduction of the original and the reconstruction space. The 2 visualizations should contain similar regions of points')
reducer = umap.UMAP()
# Learn UMAP reduction on the original space
original_umap = reducer.fit_transform(np.transpose(df.values))
# Plot UMAP visualization of the original space
plt.figure(figsize=(20, 10))
plt.scatter(original_umap[:, 0], original_umap[:, 1], c=[sns.color_palette()[0] ])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the Original space', fontsize=fontsize);
plt.show()

# Learn UMAP reduction on the reconstructed space
reconstuction_umap = reducer.fit_transform(np.transpose(outputs_val))
# Plot UMAP visualization of the reconstructed space
plt.figure(figsize=(20, 10))
plt.scatter(reconstuction_umap[:, 0], reconstuction_umap[:, 1], c=[sns.color_palette()[1] ])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the Reconstruction space', fontsize=fontsize);
plt.show()


max_clusters = 20
colors = cm.rainbow(np.linspace(0, 1, max_clusters))
print('\nSelecting the best number of cluster components [1-20] using BIC.')
print('Lower Bayesian Information Criterion (BIC) is beter.\n')

# Learn UMAP reduction on the reduced space
reduced_umap = reducer.fit_transform(reduced_val)

best_num_clusters = None
best_clustering = None
best_BIC = float("inf")

for components in range(2, max_clusters + 1, 1):    
    # Learn GMM
    g = mixture.GaussianMixture(n_components=components)
    clustering = g.fit(reduced_umap)
    cluster_labels = clustering.predict(reduced_umap)
    bic = g.bic(reduced_umap)
    print('Using %d clusters - BIC: %f' % (components, bic))
    if bic < best_BIC:
        best_BIC = bic
        best_clustering = cluster_labels
        best_num_clusters = components
    else:
        break
print('Best BIC=%f achieved with %d clusters.\n' % (best_BIC, best_num_clusters))

print('V4: A 2D UMAP reduction of the condensed space (middle layer/encoder output) and its best cluster assignment.')
# Plot UMAP visualization of the reduced space and cluster assignments
plt.figure(figsize=(20, 10))
plt.scatter(reduced_umap[:, 0], reduced_umap[:, 1], c=best_clustering, cmap=plt.cm.tab20)
plt.colorbar(cmap=sns.color_palette(), boundaries=np.arange(best_num_clusters + 1)-0.5).set_ticks(np.arange(best_num_clusters))
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the condensed space and cluster assignments for %d clusters' % best_num_clusters, fontsize=fontsize);
plt.show()


print('V5: A 2D UMAP reduction of the original space annotated with the best condensed UMAP cluster assignment.')
# Plot UMAP visualization of the original space and cluster assignments
plt.figure(figsize=(20, 10))
#     for label, embed, cluster_label in zip(labels, reduced_umap, cluster_labels):
plt.scatter(original_umap[:, 0], original_umap[:, 1], c=best_clustering, cmap=plt.cm.tab20)
#     plt.text(embed[0] - 0.02, embed[1] + 0.02, label, fontsize=14)
plt.colorbar(cmap=sns.color_palette(), boundaries=np.arange(best_num_clusters + 1)-0.5).set_ticks(np.arange(best_num_clusters))
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the original space with cluster assignments learned from the reduced space', fontsize=fontsize);
plt.show()



print('V6: A 2D t-SNE reduction of the original space annotated with the best condensed UMAP cluster assignment.')
tsne_fit = TSNE(n_components=2).fit_transform(np.transpose(df.values))
plt.figure(figsize=(20, 10))
# for label, embed in zip(labels, tsne_fit):
plt.scatter(tsne_fit[:, 0], tsne_fit[:, 1], c=best_clustering, cmap=plt.cm.tab20)
#     plt.text(embed[0] - 0.02, embed[1] + 0.02, label, fontsize=14)
plt.colorbar(cmap=sns.color_palette(), boundaries=np.arange(best_num_clusters + 1)-0.5).set_ticks(np.arange(best_num_clusters))
plt.gca().set_aspect('equal', 'datalim')
plt.title('T-SNE projection of the Input space', fontsize=fontsize);
plt.show()

````

### Model Description

My model is a stacked autoencoder supportin variable layer size.
Autoencoders are able to learn representations for a set of data in a unsupervised manner.
The stacked autoencoder consists of two smaller multilayer networks, an encoder that takes the input and compresses it into a progressively smaller and smaller latent representation.
Contrastingly, the decoder network takes the representation created by the decoders and maps it to progressively larger representations with the final output being the same size as the original representation fed to the encoder network.
Both the encoder and the decoder used three layers but in my implementation this can be changed by modifying a single line of code (a list containing the encoder sizes). In order to calculate the hidden representation for each hidden layers we multiply the current representation $r$ with a $c*t$ weight matrix W and add to it a $t$-dimensional bias vector, where $c$ are the number of dimensions of the current representation and $t$ is the number of dimensions of the target representation. Then, we create the new representation $h$ by utizing an activation function $\sigma$, thus $h = \sigma(Wr + b)$. Last, we update the current representation so that $r = h$. Usually, for autonencoders non-linear activation functions such as relu, tanh, and sigmoid are used. In this case I've used relus for all the internal layers and a sigmoid for the final output since the data $\in [0, 1]$. 

The combined network is trained so that the decoder output resembles as much as possible the original input.
This is achieved by defining a loss function over the input and the decoder output.
The network can be trained using any variant of gradient descent by minimising the loss function and back propagating through the network to update its parameters.
So, the stacked autoencoder is trained to learn a network that maps its input into a smaller hidden representation, which it then uses to reconstruct the original input.
Common loss functions that are used for autoencoders are cross-entropy, binary cross-entropy, and mean square eror.
Cross-entropy assumes that the output dimensions are a probability distribution that sums up to 1, which definitely does not seem to be the case for the given dataset so it not a good choice.
Binary cross-entropy, that each of the outputs is independent and a seperate probability distribution, so each of the outputs is in the range [0, 1]. The given data are in the range [0, 1] but there is no strong indication that they are probabilities, so I avoided using this loss. In general, this is popular loss used in various problems such as multilabel classification.
Finally, the mean square error (MSE) which is the chosen loss function measures the average squared difference between the estimated values and the actual values. In this case, we measure the average squared difference between the reconstruction values and the actual values, thus for an input vector I and an output vector O, $MSE = \frac{1}{N}\sum_{i=1}^{N}(I_i - O_i)$. Similarly, we can calculate the MSE over a batch of datapoints, or even over the whole dataset by simply averaging the MSEs of each of the datapoints.
